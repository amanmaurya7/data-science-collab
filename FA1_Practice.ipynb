```python
df['Result'] = df['Total_Score'].apply(lambda x: 'Pass' if x>=50 else 'Fail')
```

```python
df['Result']
```

```python
def performance(x):
    if x>=80: return 'Excellent'
    elif x>=60: return 'Good'
    else: return 'Average'

df['Performance'] = df['Total_Score'].apply(performance)
```

```python
df['Performance']
```

```python
Q1=numeric_df.quantile(0.25)
Q3=numeric_df.quantile(0.75)
IQR = Q3-Q1
lower = Q1-1.5*IQR
upper = Q3+1.5*IQR

df_clean=df[~((numeric_df<lower) | (numeric_df>upper)).any(axis=1)]
```
